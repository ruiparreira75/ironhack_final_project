```python
#converted prices from cents to dolares 
df_spy["price"] = df_spy["price"] / 100
df_spy["initialprice"] = df_spy["initialprice"] / 100
```

# cleaning owners collumn 

```python
#spliting collums to min and max 
split_owners = df_spy["owners_clean"].str.split(" .. ", expand=True)
split_owners.columns = ["owners_min", "owners_max"]
df_spy = df_spy.join(split_owners)
```

```python
df_spy["owners_min"] = pd.to_numeric(df_spy["owners_min"], errors="coerce")
df_spy["owners_max"] = pd.to_numeric(df_spy["owners_max"], errors="coerce")
```

```python
#creating collumn with average owners 
df_spy["owners_mid"] = (df_spy["owners_min"] + df_spy["owners_max"]) // 2
```

```python
#dropping redudant collums 
df_spy = df_spy.drop(columns=["owners", "owners_min", "owners_max"])
```

```python
#dropping redudant collums 
df_spy = df_spy.drop(columns=["owners_clean"])
```

```python
df_spy.head() 
```

```python
#rename collumn back to owners 
df_spy = df_spy.rename(columns={"owners_mid": "owners"})
```

```python
#quick check test for owners > 5000000 
df_spy[df_spy["owners"] > 5000000]
```

```python
df_spy.info() 
```

```python
df_spy["userscore"].unique()
```

```python
# Dropping userscore because is always zero. 
df_spy = df_spy.drop(columns=["userscore"])
```

```python
df_spy.head() 
```

```python
# converting collumns from minutes to hours (average_forever,average_2weeks,median_forever,	median_2weeks
```

```python
df_spy["avg_playtime_hr"] = (df_spy["average_forever"] / 60).round(2)
df_spy["med_playtime_hr"] = (df_spy["median_forever"] / 60).round(2)
df_spy["avg_2weeks_hr"] = (df_spy["average_2weeks"] / 60).round(2)
df_spy["med_2weeks_hr"] = (df_spy["median_2weeks"] / 60).round(2)
```

```python
df_spy.head() 
```

```python
#dropping the old minute collumns playtime 
df_spy = df_spy.drop(columns=[ "average_forever", "average_2weeks", "median_forever", "median_2weeks", "avg_playtime_2weeks_hr", "med_playtime_2weeks_hr"])
```

```python
df_spy.head() 
```

```python
#creating collums for total reviews (positive+negative) and review ratio 

df_spy["total_reviews"] = df_spy["positive"] + df_spy["negative"]
df_spy["review_ratio"] = (df_spy["positive"] / df_spy["total_reviews"]).round(2)

# Reorder columns to insert after 'negative'
cols = list(df_spy.columns)
insert_at = cols.index("negative") + 1

for new_col in ["total_reviews", "review_ratio"]:
    cols.insert(insert_at, cols.pop(cols.index(new_col)))
    insert_at += 1

# Apply reordered columns
df_spy = df_spy[cols]
```
